## I) Installing required packages

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
import time

import pandas as pd
```

## II) To obtain the total number of elections needed, the following are required

```python
 # Set before entering the website  #
service = Service(executable_path='chromedriver-win64/chromedriver.exe')
options = webdriver.ChromeOptions()
options.add_argument( '--ignore-certificate-errors' )
options.add_argument( '--incognito' )  # To enter incognito mode

driver = webdriver.Chrome(service = service, options = options)
driver.maximize_window()
```

```python
# Accessing the web page
enlace_base = "https://infogob.jne.gob.pe/Eleccion"
driver.set_page_load_timeout( 20 )
    
web = driver.get(enlace_base)
time.sleep(15)
```

```python
# To make a selection in filter "Type of process" = Presidential elections
filtro_proceso = driver.find_element(By.XPATH, '//*[@id="section"]/div[2]/div[2]/div[2]/div[1]/div')
filtro_proceso.click()
time.sleep(2)
    
# Move down in the drop-down list three times
for _ in range(1):
    ActionChains(driver).send_keys(Keys.ARROW_DOWN).perform()
    time.sleep(2)
    
# Select the currently highlighted option
ActionChains(driver).send_keys(Keys.ENTER).perform()
time.sleep(2)
```

```python
# To get total of options in "Choice" filter
filtro_eleccion = driver.find_element(By.XPATH,'//*[@id="IdEleccion"]')
select = Select(filtro_eleccion)
all_eleccion = select.options
total_eleccion = len(all_eleccion)

print(f"Total de elementos en la lista desplegable: {total_eleccion}")

total_eleccion
```

## III) Starting searching on the web

```python
# Loop configuration
service = Service(executable_path='chromedriver-win64/chromedriver.exe')
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)
driver.maximize_window()

enlace_base = "https://infogob.jne.gob.pe/Eleccion"
```

```python
# Place for the tables obtained
tables_list = []


i = total_eleccion + 1

for k in range(1,i):
    time.sleep(2)
    # Set before entering the website
    driver.get(enlace_base)
    time.sleep(2)

    # To make a selection in the filter "Process type".
    filtro_proceso = driver.find_element(By.XPATH, '//*[@id="section"]/div[2]/div[2]/div[2]/div[1]/div')
    filtro_proceso.click()
    time.sleep(2)

    # Move down in the drop-down list once
    ActionChains(driver).send_keys(Keys.ARROW_DOWN).perform()
    time.sleep(2)

    # Select the currently highlighted option
    ActionChains(driver).send_keys(Keys.ENTER).perform()
    time.sleep(2)

    # To make a selection in the "Choice" filter
    filtro_eleccion = driver.find_element(By.XPATH, '//*[@id="section"]/div[2]/div[2]/div[2]/div[2]/div')
    filtro_eleccion.click()
    time.sleep(5)

    # Move down in the drop-down list k times
    for _ in range(k):
        ActionChains(driver).send_keys(Keys.ARROW_DOWN).perform()
        time.sleep(5) # lo cambié de 0.5 a 2 s.

    # Select the currently highlighted option
    ActionChains(driver).send_keys(Keys.ENTER).perform()
    time.sleep(5) # lo cambié de 0.5 a 2 s.
    
    # To enter to see the data of the selected election
    boton_datos = driver.find_element(By.XPATH, '//*[@id="btnVerDatos"]')
    boton_datos.click()
    time.sleep(10)
    
    # To enter to view the results table of the selected election
    resultado_datos = driver.find_element(By.XPATH, '//*[@id="section"]/div[2]/div[3]/div[1]/ul/li[2]/a')
    resultado_datos.click()
    time.sleep(2)
    
    # Read the html text of the table.
    table_path = driver.find_element(By.XPATH, '/html/body/div[1]/section/div[2]/div[3]/div[3]/div/div/div/div[1]/div[2]/div[2]')
    table_html = table_path.get_attribute('innerHTML')
    table_data = pd.read_html(table_html)[0]
    
    # Join the obtained table in the table list.
    tables_list.append(table_data)
    time.sleep(2)
    
    # Print the table of the current iteration.
    print(f"Table Data for Iteration {k}:\n")
    print(table_data)
    print("\n" + "=" * 50 + "\n")
    
    time.sleep(2)

# Close the browser at the end of the loop.
driver.quit()
```

Editando la biblioteca de datos

```python
dfs = []
for i, table_data in enumerate(tables_list, start=1):
    df = pd.DataFrame(table_data)
    df['Número de Tabla'] = i
    dfs.append(df)

# Concatenar todos los DataFrames en uno solo
final_df = pd.concat(dfs, ignore_index=True)

# Mostrar el DataFrame final
print(final_df)
```

```python
from selenium import webdriver
service = Service(executable_path='chromedriver-win64/chromedriver.exe')
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service = service, options = options)
driver.maximize_window()

enlace_base = "https://infogob.jne.gob.pe/Eleccion"
driver.set_page_load_timeout( 20 ) 
driver.get(enlace_base)

wait = WebDriverWait(driver, 10)  # 10 seconds wait
button = wait.until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/section/div[2]/div[2]/div[2]/div[1]/div/div[1]')))
button.click()
elecciones = wait.until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/section/div[2]/div[2]/div[2]/div[1]/div/div[2]/div[2]')))
elecciones.click()
tipo = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="EleccionView"]')))
tipo.click()
```

```python
numero_iteraciones = total_eleccion - 1 # Puedes ajustar este valor según sea necesario

# Lista para almacenar los resultados
resultados = []

# Bucle para iterar sobre diferentes valores de "n"
for n in range(2, 2 + numero_iteraciones):
    # Construir el XPath con el valor actual de "n"
    xpath = f'/html/body/div[1]/section/div[2]/div[2]/div[2]/div[2]/div/div[2]/div[{n}]'

    # Encontrar el elemento y extraer el texto
    try:
        filtro_proceso = driver.find_element(By.XPATH, xpath).text
        resultados.append(filtro_proceso)
    except Exception as e:
        print(f"No se pudo encontrar el elemento para n={n}. Error: {e}")

# Imprimir los resultados
print(resultados)

# Cerrar el navegador al finalizar
driver.quit()
```

```python
data = {
    "Elecciones": resultados,
    "Número de Tabla": list(range(1, len(resultados) + 1))
}

# Crear el DataFrame
df = pd.DataFrame(data)
df
```

```python
Procesos_electorales = pd.merge(final_df, df, on="Número de Tabla")
print(Procesos_electorales)
```

```python
desired_columns = ['Elecciones', 'ORGANIZACIÓN POLÍTICA', 'TOTAL VOTOS']

# Create a new DataFrame with the selected columns
Elections = Procesos_electorales[desired_columns]

# Reorganize the order of columns as per your specifications
Elections = Elections[['Elecciones', 'ORGANIZACIÓN POLÍTICA', 'TOTAL VOTOS']]

# Export the resulting DataFrame to an Excel file
Elections.to_excel('Elections_2021_1931.xlsx', index=False)
```
