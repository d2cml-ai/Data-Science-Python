```python
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Initialize the WebDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
driver.maximize_window()

# URL
url = 'https://finance.yahoo.com/'
driver.get(url)

# Wait for the navigation menu to load
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "ybar-navigation")))

# Click on the third item in the navigation menu
market = driver.find_element(By.XPATH, '//*[@id="ybar-navigation"]/div/ul/li[3]/a')
driver.execute_script("arguments[0].scrollIntoView({block: 'center'});", market)  # Center scroll
time.sleep(1)  # Short pause to ensure scrolling
market.click()

# Wait for the new page to load
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.TAG_NAME, "body")))

# Scroll gradually to find the S&P 500 link
company = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.XPATH, '//a[@aria-label="S&P 500"]'))
)
driver.execute_script("arguments[0].scrollIntoView({block: 'center'});", company)  # Center scroll on the element
company.click()

# Wait for the S&P 500 page to fully load
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.TAG_NAME, "body")))

# Scroll down
driver.execute_script("window.scrollBy(0, 200);")

# Wait for the "Historical Data" button to be clickable
historical_data = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.CSS_SELECTOR, '[title="Historical Data"]'))
)
historical_data.click()

# Wait for the table to appear
datah = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.CLASS_NAME, "table-container"))
)

# Extract all rows (excluding the first header row)
rows = datah.find_elements(By.TAG_NAME, "tr")[1:]  # Skipping the first row

# Extract data (td elements)
data = []
for row in rows:
    cols = row.find_elements(By.TAG_NAME, "td")
    data.append([col.text.strip() for col in cols])

# Create the DataFrame without headers
df = pd.DataFrame(data)

# Display the DataFrame
print(df)

# Close the browser
driver.quit()
```
